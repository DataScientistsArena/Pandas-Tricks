```python
import numpy as np
import pandas as pd

from itertools import product

datecols = ['year', 'month', 'day']

df = pd.DataFrame(list(product([2017, 2016], [1, 2], [1, 2, 3])),
                  columns=datecols)

df['data'] = np.random.randn(len(df))

df.index = pd.to_datetime(df[datecols])
df
```
