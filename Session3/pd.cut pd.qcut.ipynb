```python

import random

# np.random.randint(low = 3,high=8,size=10)
df = pd.DataFrame({"nums":random.sample(range(1, 100), 15)})
```

```python
df["bin"] = pd.cut(df["nums"], bins = [0, 15, 50, 80, 99])

df["bin_cat"] = pd.cut(df["nums"], bins = [0, 15, 50, 80, 99],labels = ["awful", "bad", "average", "good"])
```

```python
df["divide_equal_4"] = pd.qcut(df["nums"], q = 4)

df["divide_equal_cat"] = pd.qcut(df["nums"], q = 4,labels = ["awful", "bad", "average", "good"])
```

```python
df["nums_cat"] = pd.qcut(df["nums"], q = 4,labels = [0,1,2,3])
```

```python
bin_labels_5 = ['Bronze', 'Silver', 'Gold', 'Platinum', 'Diamond']
df['nums_cat2'] = pd.qcut(df['nums'],
                              q=[0, .1, .4, .5, .8, 1],
                              labels=bin_labels_5)
df
```
